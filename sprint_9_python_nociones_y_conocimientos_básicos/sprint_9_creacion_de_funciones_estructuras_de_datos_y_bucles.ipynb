## 1.1. Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en: 🔗 Índice de masa corporal IMC que es y como se calcula.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

```python
def pedir_dato_float(pedido: str) -> float:
    """
    Solicita al usuario una cadena de texto por consola.

    Args:
        pedido: Mensaje pidiendo el valor al usuario.
    
    Returns:
        El valor numérico introducido convertido en tipo float.
    """
    try:
        return float(input(pedido).replace(",", "."))
    except:
        raise ValueError ("El valor introducido debe ser un número.")
    
def validar_numero_positivo(valor: float, nombre_campo: str) -> None:
    """
    Valida que el valor de cada campo sea mayor que 0.

    Args:
        valor: el valor numérico
        nombre_campo: el campo a que se le atribuye el valor evaluado.
    """
    if valor <= 0:
        raise ValueError(f"El valor de {nombre_campo} debe ser mayor que 0.")

def calculo_imc(peso: float, altura:float) -> float:
    """
    Calcula el índice de masa corporal:
    peso entre el cuadrado de la altura.

    Args:
        peso: El valor en tipo float.
        altura: El valor en tipo float.

    Returns:
        El resultado del cálculo redondeado a 2 decimales
    """
    return round(peso / (altura ** 2), 2)

def clasificacion_imc(IMC: float) -> str:
    """
    Categorización del IMC introducido según la estandarización propuesta
    por la OMS.

    Args:
        imc: El valor del índice de masa corporal .
    
    Returns:
        La categoría según el rango en que se encuentre el imc.
    """
    if IMC < 18.5:
        return "bajo peso"
    elif IMC < 25:
        return "peso normal"
    elif IMC < 30:
        return "sobrepeso"
    else:
        return "obesidad"
    
def mostrar_resultado(imc: float, categoria_peso: str) -> None:
    """ 
    Imprime el resultado del IMC y su categoría.

    Args:
        imc: IMC calculado en tipo float.
        categoria_peso: la categoría resultante de la clasificación.
    """
    print(f'Tu IMC es {imc}, tienes {categoria_peso}.')

def calculadora_imc() -> None:
    """ 
    Bucle principal de la calculadora de IMC.

    Solicita peso y altura al usuario, valida que sean
    valores positivos, calcula el IMC y muestra el resultado
    con su categoría según los rangos de la OMS.
    Repite si los datos introducidos no son válidos.
    """
    while True:
        try:
            peso = pedir_dato_float("Introduce tu peso en kg: ")
            altura = pedir_dato_float("Introduce tu altura en metros: ")
            validar_numero_positivo(peso, "peso")
            validar_numero_positivo(altura, "altura")
            print(f'Siendo tu peso {peso} kg. y tu altura {altura} m. : ')
            imc = calculo_imc(peso, altura)
            categoria_peso = clasificacion_imc(imc)
            mostrar_resultado(imc, categoria_peso)
            break
        except (TypeError, ValueError) as error:
            print(f"{type(error).__name__}: {error}")
            
```

## 1.2. Convertidor de temperatures

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
A_CELSIUS = {
    'CELSIUS': lambda t:t,
    'FAHRENHEIT': lambda t: (t - 32) * 5/9,
    'KELVIN': lambda t: t - 273.15
}

DESDE_CELSIUS = {
    'CELSIUS': lambda t: t,
    'FAHRENHEIT': lambda t: t * 9/5 + 32,
    'KELVIN': lambda t: t + 273.15
}
UNIDADES_VALIDAS = ['CELSIUS', 'FAHRENHEIT', 'KELVIN']

def pedir_dato_texto(mensaje:str) -> str:
    """
    Solicita al usuario un texto por consola.

    Args:
        mensaje: texto de pedido que se muestra al usuario.
    Returns:
        El texto recibido sin espacios iniciales ni finales.
    """
    return input(mensaje).strip()

def validar_unidad_temperatura(unidad:str) -> str:
    """
    Valida que la unidad de temperatura sea una de las aceptadas.

    Convierte la entrada a mayúsculas antes de compararla,
    así 'celsius', 'Celsius' y 'CELSIUS' son equivalentes.

    Args:
        unidad: Cadena con la unidad a evaluar.

    Returns:
        La unidad en mayúsculas si es válida.

    Raises:
        ValueError: Si la unidad no pertenece a UNIDADES_VALIDAS.
    """
    unidad_pasada = unidad.upper()
    if unidad_pasada not in UNIDADES_VALIDAS:
        raise ValueError(f"Unidad no válida: {unidad_pasada}. Elegir entre: {UNIDADES_VALIDAS}")
    return unidad_pasada

def operacion_convertir(temperatura_valor: float, temperatura_unidad: str) -> dict[str, float]:
    """
    Convierte una temperatura a las demás unidades disponibles.

    Si la unidad ingresada es CELSIUS, aplica directamente las fórmulas de conversión.
    En caso contrario, realiza una conversión intermedia a CELSIUS para luego convertirlo
    al resto.

    Args:
        temperatura_valor: valor numérico de la temperatura a convertir.
        temperatura_unidad: unidad de la temperatura de entrada.

    Returns:
        conversiones: diccionario con los resultados de la conversion, con las otra unidades
        como claves y sus valores numéricos (con 2 decimales) como valores.
    """
    if temperatura_unidad == 'CELSIUS':
        conversiones = {
            key: round(formula(temperatura_valor), 2)
            for key, formula in DESDE_CELSIUS.items()
            if key != temperatura_unidad
        }
    else:
        conversiones_a_celsius= {
            key: round(formula(temperatura_valor), 2)
            for key, formula in A_CELSIUS.items()
        }
        conversiones = {            
            key: round(formula(conversiones_a_celsius[temperatura_unidad]), 2)
            for key, formula in DESDE_CELSIUS.items()
            if key != temperatura_unidad
        }
    return conversiones

def resultados_diccionario(un_diccionario: dict[str, float]) -> None:
    """
    Imprime por consola los pares clave-valor de un diccionario.

    Args:
        un_diccionario: el diccionario cuyas entradas se quieren mostrar.
    """
    for clave, valor in un_diccionario.items():
        print(f'{clave}: {valor}')

def conversor_temperatura() -> None:
    """
    Bucle principal del conversor de temperaturas. Reutiliza pedir_dato_float()
    para el valor numérico.

    Solicita al usuario un valor numérico y una unidad de temperatura,
    realiza la conversión al resto de unidades y muestra los resultados.
    Repite si los datos introducidos no son válidos.
    """
    while True:
        try:
            valor_temp = pedir_dato_float("Introduzca el número de grados que quiere convertir: ")
            unidad_temp = validar_unidad_temperatura(pedir_dato_texto(f"Introduzca la unidad entre {UNIDADES_VALIDAS}: "))
            resultado_conversiones = operacion_convertir(valor_temp, unidad_temp)
            print(f'Para {valor_temp} grados {unidad_temp}: ')
            resultados_diccionario(resultado_conversiones)
            break
        except (TypeError, ValueError) as error:
            print(f'{type(error).__name__}: {error}')
```

## 1.3. Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
from collections import Counter
from statistics import mean
import pprint
import string
```

```python
def validar_texto(texto: str) -> None:
    """Valida que el texto introducido sea procesable.

    Comprueba que no esté vacío ni compuesto exclusivamente por números.

    Args:
        texto: Cadena de texto a validar.

    Raises:
        ValueError: Si el texto está vacío o contiene solo espacios.
        ValueError: Si todas las palabras del texto son números.
    """
    if not texto or not texto.strip():
        raise ValueError ("El texto no puede estar vacío o solo contener espacios.")
    if all(word.isdigit() for word in texto.strip().split()):
        raise ValueError ("El texto no puede contener exclusivamente números.")

def contar_palabras(texto: str) -> dict[str, int]:
    """
    Cuenta las ocurrencias de cada palabra en el texto.

    Normaliza a minúsculas y elimina la puntuación antes de contar.

    Args:
        texto: Cadena de texto a procesar.

    Returns:
        Diccionario con las palabras como claves y su número de
        apariciones como valores.
    """
    texto = texto.lower()
    for simbolo in string.punctuation + '¡¿':
        texto = texto.replace(simbolo, " ")
    return dict(Counter(texto.split()))

def promedio_largo_palabras(diccio_conteo: dict) -> None:
    """
    Calcula e imprime el promedio de caracteres por palabra única.
    Reutiliza contar_palabras() para obtener las palabras del texto ya separadas
    y calcula la media de la longitud de estas.

    Args:
        texto: Cadena de texto a procesar.

    Note:
        El promedio se calcula sobre palabras únicas, no sobre
        todas las ocurrencias.
    """    
    promedio_largo_palabra = round(mean(map(len, diccio_conteo.keys())),2)
    return promedio_largo_palabra

def resultados_conteo(conteo_final: dict[str, int], promedio: float) -> None:
    """
    Imprime por consola los pares clave-valor de un diccionario y el promedio
    del largo de las palabras en este.

    Args:
        conteo_final: el diccionario cuyas entradas se quieren mostrar.
        promedio: el valor del promedio del largo de las claves.
    """
    pprint.pprint(conteo_final)
    print(f'El promedio del largo de las palabras es {promedio} letras.')

def contador_frecuencias_palabra() -> None:
    """
    Bucle principal del contador de frecuencias de palabras.

    Solicita un texto al usuario, lo valida, muestra el conteo
    de cada palabra y el promedio de caracteres por palabra única.
    Repite si el texto introducido no es válido.
    """
    while True:
        try:
            texto = pedir_dato_texto('Introduce tu texto aquí: ')
            validar_texto(texto)
            conteo = contar_palabras(texto) 
            promedio_lp = promedio_largo_palabras(conteo)
            resultados_conteo(conteo, promedio_lp)
            break
        except (TypeError, ValueError) as error:
            print(f'{type(error).__name__}: {error}')
            
```

## 1.4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.


```python
def inversor_diccionarios(diccio: dict) -> dict:
    """
    Intercambia claves y valores de un diccionario.

    Si un valor aparece más de una vez, agrupa las claves originales
    en una lista e imprime un aviso por cada valor duplicado encontrado.

    Args:
        diccio: Diccionario no vacío a invertir.

    Returns:
        Nuevo diccionario con los valores originales como claves.
        Si hay duplicados, el valor correspondiente es una lista
        con todas las claves que lo tenían.

    Raises:
        TypeError: Si el argumento no es un diccionario.
        ValueError: Si el diccionario está vacío.
    """
    if not isinstance(diccio, dict):
        raise TypeError ("El argumento debe ser un diccionario.")
    if not diccio:
        raise ValueError ("El diccionario no puede estar vacío.")
    
    nuevo_diccio = {}
    duplicados_avisados = set()
    for clave, valor in diccio.items():
        if valor not in nuevo_diccio:
            nuevo_diccio[valor] = clave
        else:
            if valor not in duplicados_avisados:
                print(f"Valor duplicado encontrado: '{valor}'")
                duplicados_avisados.add(valor)
            if not isinstance(nuevo_diccio[valor], list):
                nuevo_diccio[valor] = [nuevo_diccio[valor]]
            nuevo_diccio[valor].append(clave)
    return nuevo_diccio
```

## 1.1. Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z.


```python
import string
from pathlib import Path
```

```python
def analizar_archivo(ruta: str) -> str:
    """
    Lee el contenido de un archivo de texto.

    Args:
        ruta: Ruta al archivo .txt a leer.

    Returns:
        El contenido completo del archivo como cadena de texto.

    Raises:
        FileNotFoundError: Si no existe ningún archivo en la ruta indicada.
    """    
    if not Path(ruta).exists():
        raise FileNotFoundError(f"No se encontró el archivo: {ruta}")
    return Path(ruta).read_text(encoding="utf-8")

def diccionario_contador_palabras(txt: str) -> dict[str, dict[str, int]]:
    """ 
    Organiza las palabras de un texto y sus frecuencias por letra inicial.

    Reutiliza validar_texto() para validar la entrada y contar_palabras()
    para la limpieza y el conteo base. Las letras sin palabras se omiten
    en el resultado final.

    Args:
        txt: Cadena de texto a procesar.

    Returns:
        Diccionario con las letras del abecedario como claves y, como valor,
        un diccionario con las palabras que empiezan por esa letra y su
        frecuencia. Las letras sin palabras no aparecen.

    Raises:
        ValueError: Si el texto está vacío o contiene solo números o espacios.
                    (propagado desde validar_texto).
    """
    validar_texto(txt)
    diccionario_base = contar_palabras(txt)
    diccionario_alfabetico = {letra: {} for letra in string.ascii_lowercase}
    for palabra, frecuencia in sorted(diccionario_base.items()):
        primera_letra = palabra[0]
        diccionario_alfabetico[primera_letra][palabra] = frecuencia

    return {key: value for key, value in diccionario_alfabetico.items() if value}

def mostrar_resultado_alfabetico(diccionario: dict[str, dict[str, int]]) -> None:
    """
    Imprime por consola un diccionario de frecuencias organizado por letra.

    Args:
        diccionario: Diccionario con letras como claves y sub-diccionarios
                     de palabras y frecuencias como valores.
    """    
    for letra, palabras in diccionario.items():
        print(f"\n--- {letra.upper()} ---")
        for palabra, frecuencia in palabras.items():
            print(f"    {palabra}: {frecuencia}")

def diccionario_alfabetico_frecuencias(ruta: str) -> None:
    """
    Función principal del analizador de texto por letra inicial.

    Lee un archivo, calcula las frecuencias de cada palabra y las
    muestra organizadas alfabéticamente.

    Args:
        ruta: Ruta al archivo .txt a analizar.
    """
    
    try: 
        texto = analizar_archivo(ruta)
        resultado = diccionario_contador_palabras(texto)
        mostrar_resultado_alfabetico(resultado)
    except (TypeError, ValueError, FileNotFoundError) as error:
        print(f'{type(error).__name__}: {error}')
```

## 2.2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir.

```python
def desempaquetador_lista(lista: list) -> list:
    """
    Desempaqueta elementos dentro de la lista que sean listas o tuplas.

    Args:
        lista: list, una lista no vacía que puede contener distintos elementos.

    Returns:
        lista_nivelada: list, todos los elementos de lista en un solo nivel.

    Raises:
        TypeError: Si el argumento no es una lista.
    """
    if not isinstance(lista, list):
        raise TypeError ("El argumento debe ser una lista")

    lista_nivelada = []
    for elemento in lista:
        if isinstance(elemento, (list, tuple)):
            lista_nivelada.extend(desempaquetador_lista(list(elemento)))
        else:
            lista_nivelada.append(elemento)
    return lista_nivelada


def conversor_float(una_lista: list) -> tuple[list, list]:
    """ 
    Convierte los elementos de una lista en tipo float.
    Reutiliza desempaquetador_lista() para obtener los elementos nivelados.

    Args:
        una_lista: una lista no vacía.

    Returns:
        tuple:
            lista_convertidas: contiene los valores convertidos a float.
            lista_no_convertidas: contiene los elementos de la lista que
                                  han dado error al intentar convertirse.
                    
    Raises:
        TypeError: Si el argumento no es una lista.
        ValueError: Si la lista está vacía.
    """
    if not isinstance(una_lista, list):
        raise TypeError ("El argumento debe ser una lista.")
    if not una_lista:
        raise ValueError ("La lista no puede estar vacía.")

    lista_base = desempaquetador_lista(una_lista)
    lista_convertidas = []
    lista_no_convertidas = []
    for item in lista_base:
        try:
            lista_convertidas.append(float(item))
        except (ValueError, TypeError):
            lista_no_convertidas.append(item)
    return lista_convertidas, lista_no_convertidas

```

```python
lista_ejemplo = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]
conversor_float(lista_ejemplo)
```

## 3.1. Generador de Contrasenyes

Explora el funcionament del mòdul random de la llibreria numpy. 🔗 Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

longitud (int): Longitud de la contrasenya

majuscules (bool = True): Si hi ha d’aparèixer majúscules

minuscules (bool = True): Si hi ha d’aparèixer minúscules

numeros (bool = True): Si hi ha d’aparèixer números

signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)


Així doncs, si executem la funció de la següent manera:

crear_contrasenya(10, True, True, True, True)

Hauríem d’obtenir un output (que hem de poder guardar) de l’estil:

9Er,5Vn8P$

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python
import numpy as np
import string
import pyperclip

def crear_password(longitud: int, majuscules: bool = True, minuscules: bool = True, numeros: bool = True, signes: bool = False) -> None:
    """ 
    Genera un contraseña aleatoria según las categorías activadas 
    y la copia en el portapapeles.

    Garantiza que aparezca al menos un carácter de cada categoría
    activada antes de rellenar el resto aleatoriamente.
    
    Args:
        longitud:   Número de caracteres de la contraseña, debe ser
                    mayor al número de categorías activadas.
        majuscules: Si se incluyen o no las letras mayúsculas (A-Z).
                    Por defecto True.
        minuscules: Si se incluyen o no las letras minúsculas (a-z).
                    Por defecto True.
        numeros:    Si se incluyen o no los dígitos (0-9).
                    Por defecto True.
        signes:     Si se incluyen o no los símbolos de puntuación.
                    Por defecto False.
    
    Raises:
        TypeError:  Si longitud no es un entero.
        ValueError: Si no hay ninguna categoría activada.
        ValueError: Si longitud es menor que el número de categorías activas.
    """
    if not isinstance(longitud, int):
        raise TypeError ("El argumento debe ser un número entero (int).")
    categorias = []
    sopa = ''

    if majuscules:
        categorias.append(string.ascii_uppercase)
        sopa += string.ascii_uppercase
    if minuscules:
        categorias.append(string.ascii_lowercase)
        sopa += string.ascii_lowercase
    if numeros:
        categorias.append(string.digits)
        sopa += string.digits
    if signes:
        categorias.append(string.punctuation)
        sopa += string.punctuation 
    if not sopa:
        raise ValueError ("Por favor, selecciona más criterios de entre los disponibles.")
    if longitud < len(categorias):
        raise ValueError(f"La longitud mínima para los criterios seleccionados es {len(categorias)}.")
    
    obligatorias = [np.random.choice(list(categoria)) for categoria in categorias]
    las_demas = list(np.random.choice(list(sopa), size=longitud - len(obligatorias), replace=True))
    password_list = obligatorias + las_demas
    np.random.shuffle(password_list)
    password = "".join(password_list)

    print("Tu nueva contraseña ha sido copiada en el portapapeles.")
    pyperclip.copy(password)
```

```python
crear_password(10, True, True, True, True)
```

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Utilitza el fitxer "historic_partits.txt"

La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)

```python
from pathlib import Path
```

```python
def resultados_procesados_fc(archivo: str) -> dict:
    """
    Procesa un archivo de texto con partidos de fútbol y calcula estadísticas.

    Lee el archivo línea por línea esperando el formato:
    equipo_local [TAB] goles_local-goles_visitante [TAB] equipo_visitante

    Args:
        archivo: Ruta al archivo .txt con el histórico de partidos.

    Returns:
        Diccionario con las siguientes claves:
        - 'goles_marcados': lista de (equipo, goles) ordenada de mayor a menor.
        - 'equipo_goleador': nombre del equipo con más goles marcados.
        - 'goles_recibidos': lista de (equipo, goles) ordenada de mayor a menor.
        - 'equipo_mas_goleado': nombre del equipo con más goles recibidos.
        - 'clasificacion_por_puntos': lista de (equipo, puntos) ordenada de mayor a menor.

    Raises:
        FileNotFoundError: Si no existe el archivo en la ruta indicada.
        TypeError: Si el argumento no es una cadena de texto. 
        ValueError: Si el archivo está vacío.   
    """
    if not Path(archivo).exists():
        raise FileNotFoundError(f"No se encontró el archivo: {archivo}. Asegúrese de poner la extensión")
    if not isinstance(archivo, (str, Path)):
        raise TypeError (f"Se esperaba str o Path, se recibió {type(archivo).__name__}.")
    if Path(archivo).stat().st_size == 0:
        raise ValueError("El archivo está vacío.")

    goles_marcados = {}
    goles_recibidos = {}
    puntos = {}

    with open(archivo, encoding='utf-8') as fuente:
        for linea in fuente:
            partido = linea.strip().split('\t')
            equipo_local = partido[0]
            marcador = partido[1].split('-')
            goles_local = int(marcador[0])
            goles_visitante = int(marcador[1])
            equipo_visitante = partido[2]

            goles_marcados[equipo_local] = goles_marcados.get(equipo_local, 0) + goles_local
            goles_marcados[equipo_visitante] = goles_marcados.get(equipo_visitante, 0) + goles_visitante
            goles_recibidos[equipo_local] = goles_recibidos.get(equipo_local, 0) + goles_visitante
            goles_recibidos[equipo_visitante] = goles_recibidos.get(equipo_visitante, 0) + goles_local
            if goles_local > goles_visitante:
                puntos[equipo_local] = puntos.get(equipo_local, 0) + 3
                puntos.setdefault(equipo_visitante, 0)
            elif goles_local < goles_visitante:
                puntos[equipo_visitante] = puntos.get(equipo_visitante, 0) + 3
                puntos.setdefault(equipo_local, 0)
            else:
                puntos[equipo_local] = puntos.get(equipo_local, 0) + 1
                puntos[equipo_visitante] = puntos.get(equipo_visitante, 0) + 1

    clasificacion_goles_marcados = sorted(goles_marcados.items(), key=lambda x: (-x[1], x[0])) 
    equipo_goleador = clasificacion_goles_marcados[0]
    clasificacion_goles_recibidos = sorted(goles_recibidos.items(), key=lambda x : (-x[1], x[0])) 
    equipo_mas_goleado = clasificacion_goles_recibidos[0]
    clasificacion_puntos = sorted(puntos.items(), key=lambda x :(-x[1], x[0]))

    return {
        'goles_marcados': clasificacion_goles_marcados,
        'equipo_goleador': equipo_goleador[0],
        'goles_recibidos': clasificacion_goles_recibidos,
        'equipo_mas_goleado': equipo_mas_goleado[0],
        'clasificacion_por_puntos': clasificacion_puntos
    }

def presentacion_resultados_fc(diccio_resultados: dict) -> None:
    print("Equipo más goleador:", diccio_resultados['equipo_goleador'])
    print("Equipo más goleado:", diccio_resultados['equipo_mas_goleado'])
    print("\n Goles marcados:")
    for equipo, goles in diccio_resultados['goles_marcados']:
        print(f"  {equipo}: {goles} goles")
    print("\n Clasificación por puntos:")
    for equipo, pts in diccio_resultados['clasificacion_por_puntos']:
        print(f"  {equipo}: {pts} pts")

def resumen_partidos_fc(archivo: str) -> None:
    try:
        presentacion_resultados_fc(resultados_procesados_fc(archivo))
    except (FileNotFoundError, TypeError, ValueError) as error:
        print(f'{type(error).__name__}: {error}')
```

```python
resumen_partidos_fc("historic partits.txt")
```
