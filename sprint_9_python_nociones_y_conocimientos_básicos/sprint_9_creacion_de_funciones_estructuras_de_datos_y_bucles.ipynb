1. Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en: 🔗 Índice de masa corporal IMC que es y como se calcula.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

```python
def calculadora_IMC() -> tuple[float, str]:
    """
    Calcula el índice de masa corporal y devuelve esta mas su clasificación:
    bajo peso, peso normal, sobrepeso u obesidad.
    """
    while True:
        try:
            # Pedimos los valores al usuario
            peso = float(input("Introduce tu peso en kg: ").replace(",","."))
            altura = float(input("Introduce tu altura en metros: ").replace(",","."))

            # Restringimos los valores a positivo y mayor que 0
            if peso > 0 and altura > 0:
                
                # Cálculo
                IMC = round(peso / (altura**2),2)

                # Categorización según peso
                if IMC < 18.5:
                    categoria_peso = "bajo peso"
                elif IMC < 25:
                    categoria_peso = "peso normal"
                elif IMC < 30:
                    categoria_peso = "sobrepeso"
                else:
                    categoria_peso = "obesidad"

                return IMC, categoria_peso
            #Si hay un error en el valor del input() se lanza el mensaje
            else:
                print("ValueError: Los valores deben ser mayores que 0.")
            
        # Si hay un error en el tipo del input() se lanza el mensaje
        except:
            print("TypeError: El valor introducido debe ser un número.")

# Ejemplo de ejecución
ejemplo = calculadora_IMC()
print(f'Tu IMC es {ejemplo[0]}, tienes {ejemplo[1]}.')

```

2. Convertidor de temperatures

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
# Definición de las unidades y sus fórmulas mediante un diccionario
CONVERSIONES = {
    ("CELSIUS",    "FAHRENHEIT"): lambda t: t * 9/5 + 32,
    ("CELSIUS",    "KELVIN"):     lambda t: t + 273.15,
    ("FAHRENHEIT", "CELSIUS"):    lambda t: (t - 32) * 5/9,
    ("FAHRENHEIT", "KELVIN"):     lambda t: (t - 32) * 5/9 + 273.15,
    ("KELVIN",     "CELSIUS"):    lambda t: t - 273.15,
    ("KELVIN",     "FAHRENHEIT"): lambda t: (t - 273.15) * 9/5 + 32,
}

# Definición de las unidades con las que trabajaremos mediante un set
UNIDADES_VALIDAS = {"CELSIUS", "FAHRENHEIT", "KELVIN"}

def conversor_temperatura() -> dict[str, float]:
    """
    Solicita al usuario el valor y la unidad de origen mediante input(),
    valida el formato de entrada y repite la solicitud hasta recibir 
    datos correctos y devuelve su conversión en las otras 2 unidades.
    
    Unidades válidas: Celsius, Fahrenheit y Kelvin.
    """

    while True:
        try:
            # Pedimos los valores al usuario

            # Pedimos la temperatura_valor, intentamos convertir y validamos
            try:
                temperatura_valor = float(input("Introduce el número de grados que quieres convertir: "))
            except ValueError:
                print("ValueError. Por favor introduzca un número.")
                continue
            
            # Pedimos la temperatura_unidad y comprobamos que es valida
            temperatura_unidad = input("Introduce la unidad ('CELSIUS', 'FAHRENHEIT' o 'KELVIN') desde la que quieres convertir: ").upper().strip()
            if temperatura_unidad not in UNIDADES_VALIDAS:
                print("ValueError: Unidad no válida. Elija entre: 'CELSIUS', 'FAHRENHEIT' o 'KELVIN'.")
                continue

            # Generación del diccionario de resultado usando las constantes en CONVERSIONES
            conversion =  {
                destino: round(formula(temperatura_valor), 2)
                    for (origen, destino), formula in CONVERSIONES.items()
                    if origen == temperatura_unidad
            }
            return conversion
            # Si hay cualquier otro error
        except Exception:
            print("Error inesperado.")
            return None

# Ejemplo de ejecución
ejemplo_conversor = conversor_temperatura()
for unidad, temperatura in ejemplo_conversor.items():
    print(f"{unidad}: {temperatura}")
```

3. Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
def contador_palabras(texto: str) -> dict[str, int]:
    """
    Cuenta las palabras en un texto y devuelve un diccionario.
    """
    #Validación de la entrada
    if not texto or not texto.strip():
        return "ValueError: El texto no puede estar vacío o solo contener espacios."
    if all(word.isdigit() for word in texto.strip().split()):
        return "ValueError: El texto no puede contener exclusivamente números."

    # Normalización a caracteres en minúscula
    texto = texto.lower()

    # Definición de símbolos a eliminar
    quitar = ",;:.\n!)(\"-_"

    # Reemplazo de signos por espacios
    for signo in quitar:
        texto = texto.replace(signo, " ")

    # División de palabras
    palabras = texto.split()

    # Conteo
    diccionario_palabras = {}
    for palabra in palabras:
        diccionario_palabras[palabra] = diccionario_palabras.get(palabra, 0) + 1

    return diccionario_palabras

# Extra
def promedio_largo_palabra(un_texto: str) -> float:
    """
    Calcula el promedio de caracteres por palabra en un texto.
    Reutiliza contador_palabras() para obtener la palabras ya limpias.
    """
    # Validación de la entrada
    if not un_texto or not un_texto.strip():
        return "ValueError: El texto no puede estar vacío o solo contener espacios."
    if all(p.isdigit() for p in un_texto.strip().split()):
        return "ValueError: El texto no puede contener exclusivamente números."
    
    # Uso de contador_palabras() para tener las palabras separadas y normalizadas
    conteo = contador_palabras(un_texto)

    # Cálculo del total de palabras en texto
    total_palabras = sum(conteo.values())

    # Cálculo de la suma total de caracteres
    total_caracteres = sum(len(palabra) * veces for palabra, veces in conteo.items())
    
    # Cálculo del promedio
    promedio = round(total_caracteres / total_palabras, 2)
    
    return promedio

# Ejemplo de ejecución
entrada = input("Introduce una frase: ")
print(contador_palabras(entrada))
print(promedio_largo_palabra(entrada))

```

4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

Fixa’t en el següents exemples per veure l’output esperat.

```python
def inversor_diccionarios(diccio: dict) -> dict:
    """
    Recibe un diccionario e intercambia sus claves y sus valores.
    En caso de repeticiones, lanza un mensaje con la clave repetida.
    """
    # Validación de la entrada
    if not isinstance(diccio, dict):
        return "TypeError: El argumento debe ser un diccionario."
    if not diccio:
        return "ValueError: El diccionario no puede estar vacío."
    
    # Creación del diccionario de salida
    nuevo_diccio = {}

    # Inversión de claves y valores
    for clave, valor in diccio.items():
        if valor not in nuevo_diccio: # Se crea la nueva clave a partir del valor
            nuevo_diccio[valor] = []
        else:
            print(f"Valor duplicado encontrado: '{valor}'") # Se imprime un mensaje si encuentra duplicado
        nuevo_diccio[valor].append(clave) # Se incluye el nuevo valor a partir de la clave

    return nuevo_diccio

# Ejemplo de ejecución
ejemplo1 = {'x':'apple', 'y':'banana', 'z':'banana'}
ejemplo2 = {}
ejemplo3 = (6,2)

print(inversor_diccionarios(ejemplo1))
```

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z.


```python
import string
from pathlib import Path

def diccionario_contador_palabras(txt: str) -> dict:
    """ 
    Organiza las palabras de un texto y sus frecuencias en un diccionario alfabético.
    Reutiliza contador_palabras() para la limpieza y creación del diccionario
    base.
    """
    # Validación de la entrada
    if not isinstance(txt, str):
        return "TypeError: El argumento debe ser un texto."
    if not txt:
        return "ValueError: El texto no puede estar vacío."
    
    # Uso de contador_palabras() para obtener las palabras separadas y normalizadas a minúsculas
    diccionario_base = contador_palabras(txt)

    # Inicialización de un diccionario alfabético para la clasificación de diccionario_base
    diccionario_alfabetico = {letra: {} for letra in string.ascii_lowercase}

    # Población del diccionario alfabético 
    # según la primera letra de cada palabra en diccionario_base
    for palabra, frecuencia in sorted(diccionario_base.items()):
        primera_letra = palabra[0]
        if primera_letra in diccionario_alfabetico:
            diccionario_alfabetico[primera_letra][palabra] = frecuencia

    # Generación del diccionario de resultados excluyendo las letras sin niguna palabra
    resultado_limpio = {key: value for key, value in diccionario_alfabetico.items() if value}

    return resultado_limpio

# Ejemplo de ejecución
archivo_txt = Path(r"tu me quieres blanca.txt").read_text(encoding="utf-8")
intento = diccionario_contador_palabras(archivo_txt)
for letra_alfabeto, palabra_diccionario in intento.items():
    print(f"\n--- {letra_alfabeto.upper()} ---")
    for palabra, frecuencia in palabra_diccionario.items():
        print(f"    {palabra}: {frecuencia}")

```

2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir.

```python
def desempaquetador_lista(lista: list) -> list:
    """
    Desempaqueta elementos dentro de la lista que sean listas o tuplas.
    """
    #Validación de la entrada
    if not isinstance(lista, list):
        return "TypeError: El argumento debe ser una lista"
    
    # Creación de la lista de salida    
    lista_nivelada = []

    # Población de lista_nivelada
    for elemento in lista:
        if isinstance(elemento, (list, tuple)):
            lista_nivelada.extend(desempaquetador_lista(list(elemento)))
        else:
            lista_nivelada.append(elemento)
    return lista_nivelada


def conversor_float(una_lista: list) -> tuple[list, list]:
    """ 
    Convierte los elementos de una lista en tipo float.
    """
    # Validación de la entrada
    if not isinstance(una_lista, list):
        return "TypeError: El argumento debe ser una lista"
    if not una_lista:
        return "ValueError: La lista no puede estar vacía"

    # Uso de desempaquetador_lista() para nivelar los elementos  
    lista_base = desempaquetador_lista(una_lista)

    # Creación de las listas de salida
    lista_convertidas = []
    lista_no_convertidas = []

    # Población de las listas de salida en base a si la conversión da error
    for item in lista_base:
        try:
            lista_convertidas.append(float(item))
        except (ValueError, TypeError):
            lista_no_convertidas.append(item)

    return lista_convertidas, lista_no_convertidas

# Ejemplo de ejecución             
lista_ejemplo = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]
lista_ejemplo2 = ()
lista_ejemplo3 = []
lista_ejemplo4 = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', (), ('2',1,1.4,'not-a-number'),[1,2,'3','3.4'],[]]

conversor_float(lista_ejemplo)

```

1. Generador de Contrasenyes

Explora el funcionament del mòdul random de la llibreria numpy. 🔗 Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

longitud (int): Longitud de la contrasenya

majuscules (bool = True): Si hi ha d’aparèixer majúscules

minuscules (bool = True): Si hi ha d’aparèixer minúscules

numeros (bool = True): Si hi ha d’aparèixer números

signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)


Així doncs, si executem la funció de la següent manera:

Hauríem d’obtenir un output (que hem de poder guardar) de l’estil:

9Er,5Vn8P$

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python
# Creo que poner todos esos parámetros en la llamada es innecesario
# dejaré como único parámetro la longitud para evitar mayores entradas y confusiones

import numpy as np
import string
import pyperclip


def crear_password(longitud: int) -> str:
    """ 
    Crea una contraseña segura, mayor a 8 caracteres e incluyendo mayúsculas,
    minúsculas, números y símbolos aleatoriamente. Tras generarla, la copia
    automáticamente en el portapapeles.
    """
    # Validación de la entrada
    if not isinstance(longitud, int):
        return "TypeError: El argumento debe ser un número entero (int)."
    if longitud <= 8:
        return "ValueError: El argumento debe ser mayor a 8."

    # Generación de las variables que incluirán los caracteres que usaremos
    mayusculas = string.ascii_uppercase
    minusculas = string.ascii_lowercase
    numeros = string.digits
    simbolos = string.punctuation  
    sopa = mayusculas + minusculas + numeros + simbolos

    # Creación del array
    sopa_array = np.array(list(sopa))

    # Selección random de los elementos en el array 
    # con la longitud introducida en la llamada y admitiendo duplicados
    seleccion = np.random.choice(sopa_array, size=longitud, replace=True)

    # Resultado unificado en un string
    password = "".join(seleccion)

    # Extra: Se copia al portapapeles una vez generada
    print("Nueva contraseña copiada al portapapeles:")
    pyperclip.copy(password)

    return password 

# Ejemplo de ejecución
largo = 10
largo1 = [4]
largo2 = "a"
largo3 = 6
print(crear_password(largo))

```

```python
def resultados_procesados_fc(archivo):
    """
    Recibe un archivo en formato .txt con resultados de partidos de fútbol
    por equipos y devuelve el número total de goles por equipo, el nombre 
    del equipo más goleador, el equipo más goleado y la clasificación total 
    por puntos.
    """
    # Validación de la entrada
    if not isinstance(archivo, str):
        return "TypeError: El argumento debe ser un texto."
    if not archivo:
        return "ValueError: El texto no puede estar vacío."
    
    # Creación de los diccionarios necesarios para la salida
    goles_marcados = {}
    goles_recibidos = {}
    puntos = {}

    # Apertura y lectura del archivo de entrada
    with open(archivo, encoding='utf-8') as fuente:
        for linea in fuente:
            partido = linea.strip().split('\t')
            equipo_local = partido[0]
            marcador = partido[1].split('-')
            goles_local = int(marcador[0])
            goles_visitante = int(marcador[1])
            equipo_visitante = partido[2]

            # Para los goles marcados
            goles_marcados[equipo_local] = goles_marcados.get(equipo_local, 0) + goles_local
            goles_marcados[equipo_visitante] = goles_marcados.get(equipo_visitante, 0) + goles_visitante

            # Para los goles recibidos
            goles_recibidos[equipo_local] = goles_recibidos.get(equipo_local, 0) + goles_visitante
            goles_recibidos[equipo_visitante] = goles_recibidos.get(equipo_visitante, 0) + goles_local

            # Para los puntos por partido
            if goles_local > goles_visitante:
                puntos[equipo_local] = puntos.get(equipo_local, 0) + 3
                puntos.setdefault(equipo_visitante, 0) # El visitante no recibe puntos pero nos aseguramos de que exista si no estaba
            elif goles_local < goles_visitante:
                puntos[equipo_visitante] = puntos.get(equipo_visitante, 0) + 3
                puntos.setdefault(equipo_local, 0)
            else:
                puntos[equipo_local] = puntos.get(equipo_local, 0) + 1
                puntos[equipo_visitante] = puntos.get(equipo_visitante, 0) + 1

    # Total de goles marcados por equipo en orden descendente
    clasificacion_goles_marcados = sorted(goles_marcados.items(), key=lambda x: (-x[1], x[0])) 
        
    # El equipo que más goles ha metido
    equipo_goleador = clasificacion_goles_marcados[0]

    # Total de goles recibidos por equipo en orden descendente
    clasificacion_goles_recibidos = sorted(goles_recibidos.items(), key=lambda x : (-x[1], x[0])) 

    # El equipo más goleado
    equipo_mas_goleado = clasificacion_goles_recibidos[0]
        
    # Clasificatoria por puntos en orden descendente
    clasificacion_puntos = sorted(puntos.items(), key=lambda x :(-x[1], x[0]))

    return {
        'goles_marcados': clasificacion_goles_marcados,
        'equipo_goleador': equipo_goleador[0],
        'goles_recibidos': clasificacion_goles_recibidos,
        'equipo_mas_goleado': equipo_mas_goleado[0],
        'clasificacion_por_puntos': clasificacion_puntos
    }

# Ejemplo de ejecución
ruta_archivo = (r'historic partits.txt')
resultados = resultados_procesados_fc(ruta_archivo)

print("Equipo más goleador:", resultados['equipo_goleador'])
print("Equipo más goleado:", resultados['equipo_mas_goleado'])
print("\n Goles marcados:")
for equipo, goles in resultados['goles_marcados']:
    print(f"  {equipo}: {goles} goles")
print("\n Clasificación por puntos:")
for equipo, pts in resultados['clasificacion_por_puntos']:
    print(f"  {equipo}: {pts} pts")
```
