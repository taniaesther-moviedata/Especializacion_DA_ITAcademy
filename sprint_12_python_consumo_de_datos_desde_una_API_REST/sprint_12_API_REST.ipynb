2. Mostra per pantalla:

La quantitat total de cada recurs.
El codi d'estat de cada petició.

```python
json_posts = posts.json()
print(f'Total de items de la consulta /posts: {len(json_posts)}')
print(f'/posts status_code: {posts.status_code}')
```

```python
json_users = users.json()
print(f'Total de items de la consulta /users: {len(json_users)}')
print(f'/users status_code: {users.status_code}')
```

```python
json_to_dos = to_dos.json()
print(f'Total de items de la consulta /todos: {len(json_to_dos)}')
print(f'/todos status_code: {to_dos.status_code}')
```

3. Fes una petició a una publicació inexistent per obtenir un error 404 i mostra el codi d'estat rebut.

```python
from requests.exceptions import HTTPError
```

```python
no_existe = "https://api.github.com/invalid"

try:
    respuesta = requests.get(no_existe)
    respuesta.raise_for_status()
except HTTPError as http_err:
    print(f"HTTP error: {http_err}")
except Exception as err:
    print(f"Other error occurred: {err}")
else:
    print("Success!")
```
